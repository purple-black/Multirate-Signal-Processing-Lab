```python
import numpy as np
import matplotlib.pyplot as plt
```

# Fs < 2f

```python
f = 11
t = np.linspace(0,0.5,200)
x1 = np.sin(2*np.pi*f*t)

Fs = 15
Ts = 1/Fs
n = np.arange(0,0.5/Ts)
nTs = n*Ts
x2 = np.sin(2*np.pi*f*nTs)
```

```python
print("signal frequency = ", f)
print("Sampling frequency = ", Fs)
print("Ts = ", Ts)
print("nTs = ", nTs)
```

```python
plt.figure(figsize = (10,8))
plt.suptitle("Sampling a 11Hz signal with Fs = 15Hz, i.e., Fs < 2f")

plt.subplot(2,2,1)
plt.plot(t,x1,label = 'SineWave of frequency 11Hz')
plt.xlabel("time")
plt.ylabel("Amplitude")


plt.subplot(2,2,2)
plt.plot(nTs,x2,'ro', label='Sample marks after resampling at Fs = 15Hz')
plt.xlabel("time")
plt.ylabel("Amplitude")


plt.subplot(2,2,3)
plt.stem(nTs, x2, 'm', label='Sample after resampling at Fs = 15Hz')
plt.xlabel("time")
plt.ylabel("Amplitude")


plt.subplot(2,2,4)
plt.plot(nTs, x2, 'g-', label='Reconstructed SineWave')
plt.xlabel("time")
plt.ylabel("Amplitude")

plt.tight_layout()
plt.show()
```
