```python
import numpy as np
import matplotlib.pyplot as plt

def fir_lpf(cutoff, num_taps):
    taps = np.zeros(num_taps)
    for n in range(num_taps):
        if n == (num_taps - 1) // 2:
            taps[n] = 2 * cutoff
        else:
            taps[n] = np.sin(2 * np.pi * cutoff * (n - (num_taps - 1) / 2)) / (np.pi * (n - (num_taps - 1) / 2))
    return taps

def apply_filter(signal, filter_taps):
    num_taps = len(filter_taps)
    filtered_signal = np.convolve(signal, filter_taps, mode='same')
    return filtered_signal

# Define the cutoff frequency and number of taps
cutoff_freq = 0.4 * np.pi
num_taps = 64

# Generate an example input signal
time = np.linspace(0, 1, 1000)
signal = np.sin(2 * np.pi * 10 * time) + np.sin(2 * np.pi * 30 * time) + np.sin(2 * np.pi * 50 * time)

# Design the FIR low-pass filter
filter_taps = fir_lpf(cutoff_freq, num_taps)

# Apply the filter to the input signal
filtered_signal = apply_filter(signal, filter_taps)

# Plot the input signal and filtered signal
plt.figure()
plt.plot(time, signal, label='Input Signal')
plt.plot(time, filtered_signal, label='Filtered Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Input Signal and Filtered Signal')
plt.legend()
plt.grid(True)
plt.show()
```
