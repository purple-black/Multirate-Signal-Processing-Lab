```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

# Fs < f

```python
f = 40
t = np.linspace(0, 0.5, 200)
x1 = np.sin(2*np.pi*f*t)
Fs = 20
Ts = 1/Fs
n = np.arange(0, 0.5/Ts)
nTs = n*Ts
x2 = np.sin(2*np.pi*f*nTs)
```

```python
plt.plot(t,x1)
plt.title("original signal")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.show()
```

```python
plt.stem(nTs, x2)
plt.title("sampled signal")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.show()
```

frequency domain 4 POINT 

```python
fft_4 = np.fft.fft(x2,4)
plt.stem(fft_4)
plt.title("4 point dft of signal")
plt.xlabel("frequency")
plt.ylabel("amplitude")
plt.show()
```

```python
fft_8 = np.fft.fft(x2,8)
plt.stem(fft_8)
plt.title("8 point dft of signal")
plt.xlabel("frequency")
plt.ylabel("amplitude")
plt.show()
```
