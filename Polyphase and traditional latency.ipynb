```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def trad_lat(filter_length):
    latency = (filter_length - 1)//2
    return latency
```

```python
def polyphase_filter_latency(num_stage, filter_length):
    downsamplerate = 4
    latencies = []
    stage_filter = filter_length
    for stage in range(num_stage):
        stage_latency = (stage_filter - 1)//2
        latencies.append(stage_latency)
        stage_filter //= downsamplerate
    return latencies
```

```python
lpf_len = 100
trad_latency = trad_lat(lpf_len)
print("traditional filter latency: ", trad_latency)

num_stage = 5
ployphase_latency = polyphase_filter_latency(num_stage, lpf_len)
print("polyphase_filter_latency: ", ployphase_latency)
```

```python
lpf_len1 = 500
trad_latency1 = trad_lat(lpf_len1)
print("traditional filter latency: ", trad_latency1)

num_stage1 = 10
ployphase_latency1 = polyphase_filter_latency(num_stage1, lpf_len1)
print("polyphase_filter_latency: ", ployphase_latency1)
```
