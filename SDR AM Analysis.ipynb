```python
import matplotlib.pyplot as plt
import numpy as np
```

Why the carrier frequency in modulation should be of a high frequency?

```python
#m(t) = Am cos 2pi fm t
#c(t) = Ac cos 2pi fc t
fm = 5
fc = 50
t = np.linspace(0,0.5,200)
mt =  np.cos(2*np.pi*fm*t)
ct =  np.cos(2*np.pi*fc*t)
ut = mt*ct
plt.subplot(2,2,1)
plt.plot(t,mt)
plt.title("message signal")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.show()
plt.subplot(2,2,2)
plt.plot(t,ct)
plt.title("carrier signal")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.show()
plt.subplot(2,2,3)
plt.plot(t,ut)
plt.title("modulated signal")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.show()
```

# when fc < fm

```python
fm = 20
fc = 5
t = np.linspace(0,0.5,200)
mt =  np.cos(2*np.pi*fm*t)
ct =  np.cos(2*np.pi*fc*t)
ut = mt*ct
plt.subplot(2,2,1)
plt.plot(t,mt)
plt.title("message signal")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.show()
plt.subplot(2,2,2)
plt.plot(t,ct)
plt.title("carrier signal")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.show()
plt.subplot(2,2,3)
plt.plot(t,ut)
plt.title("modulated signal")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.show()
```
